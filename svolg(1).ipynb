```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")

# Physical system
m = 1    # [kg]     masse
k = 1    # [kg s^2] fj√¶rkonstant
b = 0.03 # [kg/]    dempekonstant

# Parameters
omega  = np.sqrt(k/m)
gamma  = b/m/2
omegap = np.sqrt(omega**2 - gamma**2)
Q = np.sqrt(k*m/b**2)
print(Q)

# Plotting
N_per = 20  # No. of periods
N = 1000

Tend = N_per*omega*2*np.pi

t = np.linspace(0, Tend, N)
x = np.zeros((1,N))

A = 1
phi = 0

x = A*np.exp(-gamma*t)*np.cos(omega*t + phi)
E = x**2

plt.figure(figsize=(15,3))
plt.plot(t, x, "b--")
plt.plot(t, E, "r--")
plt.xlabel("t")
plt.ylabel("x")
```
